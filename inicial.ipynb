```python
import pandas as pd
import matplotlib.pyplot as plt

# Vamos supor que 'Divida_Governos_Central' seja o nome do seu DataFrame
# Primeiro, selecione apenas as colunas que contÃªm os anos
anos = Divida_Governos_Central.columns[2:]

# SuavizaÃ§Ã£o dos dados: calculando a mÃ©dia mÃ³vel de 10 anos
media_movel = Divida_Governos_Central[anos].rolling(window=10, min_periods=1, axis=1).mean()

# Agrupamento de dados: agrupando os anos de 10 em 10
grupos_de_dez_anos = [anos[i:i+10] for i in range(0, len(anos), 10)]
media_movel_por_grupo = pd.DataFrame({f"{grupo[0]}-{grupo[-1]}": media_movel[grupo].mean(axis=1) for grupo in grupos_de_dez_anos})

# Plotando a suavizaÃ§Ã£o dos dados
plt.figure(figsize=(12, 6))
for coluna in media_movel_por_grupo.columns:
    plt.plot(media_movel_por_grupo.index, media_movel_por_grupo[coluna], label=coluna)
plt.title('SuavizaÃ§Ã£o dos Dados ao Longo dos Anos (MÃ©dia MÃ³vel de 10 Anos)')
plt.xlabel('Ano')
plt.ylabel('Valor MÃ©dio')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(True)
plt.tight_layout()
plt.show()

# Contagem de Valores NÃ£o Nulos por Ano
num_valores_nao_nulos_por_ano = Divida_Governos_Central[anos].notnull().sum()
plt.figure(figsize=(10, 6))
num_valores_nao_nulos_por_ano.plot(kind='bar', color='skyblue')
plt.title('NÃºmero de Valores NÃ£o Nulos por Ano')
plt.xlabel('Ano')
plt.ylabel('NÃºmero de Valores NÃ£o Nulos')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Vamos supor que 'Divida_Governos_Central' seja o nome do seu DataFrame
# Selecione apenas as colunas que contÃªm os anos
anos = Divida_Governos_Central.columns[2:]

# Agrupe os anos em intervalos de 10 anos
anos_agrupados = [anos[i:i+10] for i in range(0, len(anos), 10)]

# Calcule a mÃ©dia dos valores para cada intervalo de 10 anos
dados_agrupados = pd.DataFrame({f"{grupo[0]}-{grupo[-1]}": Divida_Governos_Central[grupo].mean(axis=1) for grupo in anos_agrupados})

# Crie um boxplot para os dados agrupados de 10 em 10 anos
plt.figure(figsize=(12, 6))
dados_agrupados.boxplot()
plt.title('Boxplot dos Dados Agrupados de 10 em 10 Anos')
plt.xlabel('Intervalo de Anos')
plt.ylabel('Valor MÃ©dio')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
